### Underfitting and Overfitting
faÃ§o uma funÃ§Ã£o para calcular o erro mÃ©dio absoluto

```python
#"def" sigifica que eu estou definindo uma funÃ§Ã£o,a essa funÃ§Ã£o eu dou o nome de "get_mae" 
# as variÃ¡veis dentro do parenteses significa as variÃ¡veis necessÃ¡rias para a entrada da funÃ§Ã£o "get_mae"

#essa funÃ§Ã£o tem como parametro de entrada : a "profundidade" da arvore "max_leaf_notes"
# os dados de treinamento "train_X" e "train_y"
# e os dados para ver a precisÃ£o do modelo "val_X" e "val_y"

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    #faÃ§o um modelo de Ã¡rvore de decisÃ£o
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    # a funÃ§Ã£o fit() treina meu modelo com os dados "train_X" e "train_y"
    model.fit(train_X, train_y)
    #faÃ§o a previsÃ£o do preÃ§o dos imoveis do meu modelo atraveis dos atributos colocados
    #em val_X, coloco essa previsÃ£o na variÃ¡vel "preds_val"
    preds_val = model.predict(val_X)
    #defino mae como erro absoluto mÃ©dio
    mae = mean_absolute_error(val_y, preds_val)
    #a funÃ§Ã£o "return()" retorna o valor de "mae" final da execuÃ§Ã£o da funÃ§Ã£o "get_mae"
    return(mae)
```

comparo os diferentes tamanhos da minha Ã¡rvore

```python
candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500]

#Em resumo, o cÃ³digo estÃ¡ calculando uma sÃ©rie de valores de MAE (Mean Absolute Error)
#para diferentes "leaf_size" e armazenando-os em um dicionÃ¡rio. Em seguida, ele seleciona o "leaf_size" com o menor MAE 
#como o melhor "tree_size" e atribui a esse valor Ã  variÃ¡vel "best_tree_size".
scores = {leaf_size: get_mae(leaf_size, train_X, val_X, train_y, val_y) for leaf_size in candidate_max_leaf_nodes}
best_tree_size = min(scores, key=scores.get)
print(best_tree_size)
```

treino um modelo usando todos os dados

```python
# um modelo de arvore de decisÃ£o estÃ¡ sendo criado e tribuido a variÃ¡vel "terceiro_model"
#"best_tree_size" Ã© o valor selecionado anteriormente como o melhor tamanho de Ã¡rvore.
#O modelo Ã© configurado com um nÃºmero mÃ¡ximo de nÃ³s folha igual a "best_tree_size"
#o argumento random_state Ã© definido como 1 para garantir que o modelo 
#seja reproduzÃ­vel, ou seja, gere resultados consistentes a cada execuÃ§Ã£o.
terceiro_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=1)

# o modelo "final_model" estÃ¡ pronto p ser treinado com os dados de 
#treinamento usando o mÃ©todo fit()
terceiro_model.fit(train_X, train_y)
from sklearn.metrics import mean_absolute_error

val_predictions_terceiro = terceiro_model.predict(X)
print(val_predictions_terceiro)

val_mae_terceiro = mean_absolute_error(y, val_predictions_terceiro)
print("MAE do terceiro modelo:", val_mae_terceiro)
```
