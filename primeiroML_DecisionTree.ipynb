```python
import pandas as pd

# ler o arquivo "train.csv" e coloca-lo na variável
dado_das_casa = pd.read_csv("train.csv")
#visualizar a tabela 
dado_das_casa.describe()
```

```python
#passo para visualizar o cabeçalho das colunas e escolher o que será predito e o que será as caracteristicas
# para a predição (target and features)
dado_das_casa.columns
# quando executado esse comando ".columns" veremos que a coluna de "SalePrice" é o nosso "target", o que
# nós queremos prever 
```

```python
# defino o meu alvo de predição "target" e chamo ele de "y"
# nessa variável eu coloco a coluna "SalePrice", que é a coluna que contem o preço das casas
y = dado_das_casa.SalePrice

#crio uma lista com o nome dos atributos (features)
feature_names = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']

# seleciono o dado correspondente dos atributos na variável "feacture_names" e chamo de "X"
# X será usado como carcterísticas para treinar meu modelo
X = dado_das_casa[feature_names]
```

```python
# eu vizualizo como está minha tabela de atributos X
print(X)
X.describe()
X.head()
```

### Decision Tree

```python
# criando a arvore de decisão
from sklearn.tree import DecisionTreeRegressor
#specify the model. 
#For model reproducibility, set a numeric value for random_state when specifying the model
primeiro_model =  DecisionTreeRegressor(random_state=1)

# Fit the model
primeiro_model.fit(X, y)

#PREDIÇÃO
# aqui eu imprimo a predição de todo o modelo
val_predictions_primeiro = primeiro_model.predict(X)
print(val_predictions_primeiro)

# aqui apenas a predição para as 5 primeiras casas
predictions2 = primeiro_model.predict(X.head())
print(predictions2)
```

```python
print("Primeiras 5 predições na amostra:", primeiro_model.predict(X.head()))
print("Valores reais dessas casas:", y.head().tolist())
```

```python
# aqui eu chamo a biblioteca para calcular o erro absoluto médio (MAE) so SEGUNDO modelo
from sklearn.metrics import mean_absolute_error

val_mae_primeiro = mean_absolute_error(y, val_predictions_primeiro)
print(val_mae_primeiro)
```

### Model Validation
#### separar os dados em dados de treinamento e dados de teste e depois faz a validação

nessa celula eu vou fazer meu segundo modelo de arvore de decição, agora separando os dados em dados de treinamento (train_X, train_y) e dados de validação (val_X, val_y)

pq fazer isso?

eu irei fazer isso pq o primeiro modelo, apesar de ter uma erro absoluto médio muito baixo, ele não condiz com a realidade do preço das casas. É um modelo treinado com todos os valores do dataset, e isso é um erro

pq isso é um erro?

pq é um modelo que se encaixa perfeitamente nesse banco de dados que eu estou usando, mas se for pegar outro banco de dados da mesma cidade, poderemos ver que o MAE do primeiro modelo é muito superior do que o MAE do segundo modelo

```python
# biblioteca para separação dos dados em dados de treino e dados de validação
from sklearn.model_selection import train_test_split


train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# esse segundo modelo eu treino ele através dos dados de treino e testo segundo o dados de teste 
# pq eu fix o split do banco de dados carai, entende as coisas vagabunda
segundo_model = DecisionTreeRegressor(random_state=1)

# treinando o segundo modelo
segundo_model.fit(train_X, train_y)
```

```python
# Predict with all validation observations
val_predictions = segundo_model.predict(val_X)

# print the top few validation predictions
print(train_y)
# print the top few actual prices from validation data
print(val_y)
```

```python
# aqui eu chamo a biblioteca para calcular o erro absoluto médio (MAE) so SEGUNDO modelo
from sklearn.metrics import mean_absolute_error

val_mae_primeiro = mean_absolute_error(y, val_predictions_primeiro)
print("MAE do primeiro modelo:", val_mae_primeiro)

val_mae_segundo = mean_absolute_error(val_y, val_predictions)
print("MAE do segundo modelo:", val_mae_segundo)
```

### Underfitting and Overfitting
faço uma função para calcular o erro médio absoluto

```python
#"def" sigifica que eu estou definindo uma função,a essa função eu dou o nome de "get_mae" 
# as variáveis dentro do parenteses significa as variáveis necessárias para a entrada da função "get_mae"

#essa função tem como parametro de entrada : a "profundidade" da arvore "max_leaf_notes"
# os dados de treinamento "train_X" e "train_y"
# e os dados para ver a precisão do modelo "val_X" e "val_y"

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    #faço um modelo de árvore de decisão
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    # a função fit() treina meu modelo com os dados "train_X" e "train_y"
    model.fit(train_X, train_y)
    #faço a previsão do preço dos imoveis do meu modelo atraveis dos atributos colocados
    #em val_X, coloco essa previsão na variável "preds_val"
    preds_val = model.predict(val_X)
    #defino mae como erro absoluto médio
    mae = mean_absolute_error(val_y, preds_val)
    #a função "return()" retorna o valor de "mae" final da execução da função "get_mae"
    return(mae)
```

comparo os diferentes tamanhos da minha árvore

```python
candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500]

#Em resumo, o código está calculando uma série de valores de MAE (Mean Absolute Error)
#para diferentes "leaf_size" e armazenando-os em um dicionário. Em seguida, ele seleciona o "leaf_size" com o menor MAE 
#como o melhor "tree_size" e atribui a esse valor à variável "best_tree_size".
scores = {leaf_size: get_mae(leaf_size, train_X, val_X, train_y, val_y) for leaf_size in candidate_max_leaf_nodes}
best_tree_size = min(scores, key=scores.get)
print(best_tree_size)
```

treino um modelo usando todos os dados

```python
# um modelo de arvore de decisão está sendo criado e tribuido a variável "terceiro_model"
#"best_tree_size" é o valor selecionado anteriormente como o melhor tamanho de árvore.
#O modelo é configurado com um número máximo de nós folha igual a "best_tree_size"
#o argumento random_state é definido como 1 para garantir que o modelo 
#seja reproduzível, ou seja, gere resultados consistentes a cada execução.
terceiro_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=1)

# o modelo "final_model" está pronto p ser treinado com os dados de 
#treinamento usando o método fit()
terceiro_model.fit(train_X, train_y)
from sklearn.metrics import mean_absolute_error

val_predictions_terceiro = terceiro_model.predict(X)
print(val_predictions_terceiro)

val_mae_terceiro = mean_absolute_error(y, val_predictions_terceiro)
print("MAE do terceiro modelo:", val_mae_terceiro)
```
